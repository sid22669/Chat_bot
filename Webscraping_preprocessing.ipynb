```python
 from selenium import webdriver
from selenium.webdriver.common.by import By
import time

# Initial URL
url = 'https://brainlox.com/courses/category/technical'

# Part 1: Collect h3_text and links
h3_text_and_links = []

try:
    driver = webdriver.Chrome()  # You will need to download the Chrome WebDriver or use a different browser
    driver.get(url)
    time.sleep(2)  # Add a delay to ensure the page loads completely

    # Find all classes within the "row" class
    row_classes = driver.find_elements(By.CLASS_NAME, 'row')

    for i, row in enumerate(row_classes):
        # Find all "single-courses-box" classes within the current "row"
        single_courses_boxes = row.find_elements(By.CLASS_NAME, 'single-courses-box')

        for j, single_box in enumerate(single_courses_boxes):
            try:
                # Find the "courses-content" class within the current "single-courses-box"
                courses_content = single_box.find_element(By.CLASS_NAME, 'courses-content')

                # Fetch the text data from the "h3" element within "courses-content"
                h3_text = courses_content.find_element(By.TAG_NAME, 'h3').text.strip()
                print(f"Element {i + 1} in Row {j + 1} - {h3_text}")

                # Find the "others-option" class within the current "single-courses-box"
                others_option = single_box.find_element(By.CLASS_NAME, 'others-option')

                # Find the "BookDemo-btn" link
                book_demo_btn = others_option.find_element(By.CLASS_NAME, 'BookDemo-btn')
                book_demo_link = book_demo_btn.get_attribute('href')

                # Add the h3_text and link to the list
                h3_text_and_links.append((h3_text, book_demo_link))

            except Exception as e_inner:
                print(f"An inner error occurred: {str(e_inner)}")

except Exception as e_outer:
    print(f"An outer error occurred in Part 1: {str(e_outer)}")

finally:
    # Close the driver instance for Part 1
    driver.quit()

# Part 2: Visit links and extract h3_text and p_text
for h3_text, link in h3_text_and_links:
    try:
        driver = webdriver.Chrome()  # You will need to download the Chrome WebDriver or use a different browser
        driver.get(link)
        time.sleep(2)  # Add a delay to ensure the page loads completely

        # Check if the link works (you can add your own validation logic here)
        current_url = driver.current_url
        if current_url != link:
            print(f"Clicked link: {current_url} (Link works)")

            # Extract data from the "courses-overview" class on the new page
            overview_section = driver.find_element(By.CLASS_NAME, 'courses-overview')

            # Extract the "h3" text
            h3_text = overview_section.find_element(By.TAG_NAME, 'h3').text.strip()

            # Extract the "p" text
            p_text = overview_section.find_element(By.TAG_NAME, 'p').text.strip()

            # Print or store the extracted data
            print("Course Name:", h3_text)
            print("Course Description:", p_text)

    except Exception as e_outer:
        print(f"An outer error occurred in Part 2: {str(e_outer)}")

    finally:
        # Close the driver instance for Part 2
        driver.quit()
```

```python
import pandas as pd
import re

# Load the CSV file into a pandas DataFrame
csv_file_path = "course_data.csv"  # Replace with the path to your CSV file
df = pd.read_csv(csv_file_path)

# Preprocess the text data
def preprocess_text(text):
    if isinstance(text, str):  # Check if the input is a string
        # Remove special characters and symbols, including emoticons
        text = re.sub(r"[^a-zA-Z0-9\s\.,;!?]", "", text)
        # Convert to lowercase
        text = text.lower()
        # Replace bullet points with spaces
        text = re.sub(r"\u2022", " ", text)
        # Tokenization (split text into words)
        tokens = text.split()
        # Join tokens back into a single string
        processed_text = " ".join(tokens)
        return processed_text
    else:
        return ""  # Return an empty string for non-string values

# Apply preprocessing to the "Course Description" column
df["Course Name"] = df["Course Name"].apply(preprocess_text)

# Display the preprocessed DataFrame
print(df)
```

```python
# Apply preprocessing to the "Course Description" column
df["Course Description"] = df["Course Description"].apply(preprocess_text)

# Display the preprocessed DataFrame
print(df)
```
